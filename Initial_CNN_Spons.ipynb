```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = pd.read_pickle('Spon_Dataset.pkl')
file = file.reset_index()
print(file.head())
```

```python
import sklearn
from sklearn.utils import shuffle
```

```python
file = shuffle(file)
print(file.head())
```

```python
from sklearn.model_selection import train_test_split

X = file['Color Plots']
y = file['Labels']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=101)
```

```python
import tensorflow
import keras
from keras.models import Sequential
from keras.wrappers.scikit_learn import KerasClassifier
from keras.layers import Dense, Dropout, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D, MaxPooling1D, Conv1D
```

```python
y_train_sex = [n[1] for n in y_train]
print(y_train_sex)
```

```python
def to_x_arr(lst):

    Sequence = np.asarray(lst)
    Sequence = np.expand_dims(Sequence, -1)

    return Sequence
def encode_labs(lst):
    rlst = []
    for item in lst:
        if item == 'Male':
            rlst.append(1)
        else :
            rlst.append(0)

    return rlst

X_train = [to_x_arr(lst) for lst in X_train]
y_train_sex = encode_labs(y_train_sex)

X_train = np.asarray(X_train)
y_train_sex = np.asarray(y_train_sex)

types_x = [type(x) for x in X_train]
types_y = [type(y) for y in y_train_sex]
print(set(types_x))
print(set(types_y))
```

```python
model = Sequential()

model.add(Conv2D(filters=64, kernel_size=3, activation='relu', input_shape= X_train[0].shape))
model.add(Dropout(0.2))
model.add(MaxPooling2D(pool_size=3))
model.add(Flatten())
model.add(Dense(25, activation='relu'))
model.add(Dense(1, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x=X_train, y=y_train_sex, batch_size=500, epochs=30)
```

```python
from sklearn.metrics import confusion_matrix

preds = model.predict(X_test)
print(confusion_matrix(y_test, preds))
```
